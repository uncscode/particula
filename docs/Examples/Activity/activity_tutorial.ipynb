# Activity Tutorial (Quick Check)

Simple two-component activity example to validate notebook execution.

```python
import numpy as np

# Define moles of water and organic components
moles = np.array([10.0, 5.0])
mole_frac = moles / np.sum(moles)

# Example activity coefficients (could come from a model)
activity_coeff = np.array([0.95, 1.05])

gamma = activity_coeff / mole_frac
print("Mole fractions (water, organic):", mole_frac)
print("Activity coefficients: water=%.4f, organic=%.4f" % (gamma[0], gamma[1]))
```
