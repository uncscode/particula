# Coagulation



```python
import numpy as np
import matplotlib.pyplot as plt

# From Particula
from particula.next.particles import  properties
from particula.next.dynamics.coagulation import brownian
```

```python
# creat radius

radius = np.logspace(-9, -1, 100)
mass = 4/3 * np.pi * radius**3 * 1e3
temperature = 298
pressure = 101325

kernel = brownian.brownian_coagulation_kernel_via_system_state(
    radius_particle=radius,
    mass_particle=mass,
    temperature=temperature,
    pressure=pressure,
)
```

```python
from particula.particle import Particle
PD_disc = Particle(
    particle_radius=radius,
    particle_number=radius/radius,
)

kernel_class = PD_disc.coagulation()
```

```python
print(kernel.shape)
print(kernel_class.shape)

fig, ax = plt.subplots()
ax.plot(radius, kernel)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Radius [m]')
ax.set_ylabel('Kernel [m^3/s]')
plt.show()

fig, ax = plt.subplots()
ax.plot(radius, kernel_class)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Radius [m]')
ax.set_ylabel('Kernel [m^3/s]')
plt.show()
```
